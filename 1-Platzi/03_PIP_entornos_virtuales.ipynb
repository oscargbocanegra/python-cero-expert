### Comandos para consultar entornos virtuales

Estos son algunos comandos útiles para consultar y gestionar tus entornos virtuales:

**Listar entornos virtuales en una carpeta específica:**
```bash
find ~/ruta/carpeta -type d -name "venv" -o -name "env" | sort
```

**Ver entornos virtuales activos y sus ubicaciones:**
```bash
which python
echo $VIRTUAL_ENV
```

**Listar los archivos de activación de entornos virtuales:**
```bash
find ~/ruta/carpeta -name "activate" | grep -E "bin/activate$" | sort
```

**Verificar paquetes instalados en el entorno virtual activo:**
```bash
pip list
```

**Crear un entorno virtual:**
```bash
python3 -m venv env-tmp
```

**Activar un entorno virtual:**
```bash
source env-tmp/bin/activate
```

**Desactivar un entorno virtual:**
```bash
source deactivate
```
