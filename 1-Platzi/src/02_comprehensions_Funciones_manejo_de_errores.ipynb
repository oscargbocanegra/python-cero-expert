## Modificando conjuntos

Los conjuntos pueden ser modificados utilizando varios métodos:

1. **add()**: Agregar un elemento al conjunto
2. **update()**: Actualizar/agregar múltiples elementos
3. **remove()**: Eliminar elementos específicos (genera error si no existe)
4. **discard()**: Eliminar elementos de forma segura (no genera error si no existe)
5. **clear()**: Eliminar todos los elementos


```python
suramerica = {'per', 'bol', 'ecu', 'bra', 'arg', 'chi', 'par', 'uru', 'ven', 'guy', 'sur', 'guf'}
print('Países de Sudamérica:', suramerica)
print('Total países:', len(suramerica))
print('¿Colombia está en el conjunto?:', 'col' in suramerica)
```

```python
# 1. Agregar elementos con add()
suramerica.add('col')
print('Países de Sudamérica:', suramerica)
print('Total países:', len(suramerica))
print('¿Colombia está en el conjunto?:', 'col' in suramerica)
```

```python
# 2. Actualizar/agregar múltiples elementos con update()
suramerica.update({'col', 'bra', 'ecu'})
print('Países de Sudamérica:', suramerica)
print('Total países:', len(suramerica))
```

```python
# 3. Eliminar elementos específicos con remove()
suramerica.add('bra')
suramerica.remove('bra')
print('Países de Sudamérica:', suramerica)
print('Total países:', len(suramerica))
print('¿Brasil está en el conjunto?:', 'bra' in suramerica)

```

```python
# 4. Eliminar elementos de forma segura con discard()
suramerica.discard('bra')  # No genera error aunque 'bra' ya no esté en el conjunto
print('Países de Sudamérica:', suramerica)
print('Total países:', len(suramerica))
print('¿Brasil está en el conjunto?:', 'bra' in suramerica)
```

```python
# 5. Eliminar todos los elementos con clear()
suramerica.clear()
print('Países de Sudamérica:', suramerica)
print('Total países:', len(suramerica))
```

## Operaciones con conjuntos

Las operaciones de conjuntos incluyen:

1. **Unión (|)**: Combina todos los elementos únicos de dos conjuntos
2. **Intersección (&)**: Obtiene elementos que están en ambos conjuntos
3. **Diferencia (-)**: Obtiene elementos que están en el primer conjunto pero no en el segundo
4. **Diferencia simétrica (^)**: Obtiene elementos que están en uno u otro conjunto, pero no en ambos


```python
# 1. Unión de conjuntos

# La unión de conjuntos combina todos los elementos únicos de dos o más conjuntos
# Se puede realizar con el operador | o el método union()

# Definimos dos conjuntos
countries = {'col', 'mex', 'bol'}
countries2 = {'per', 'bra'}

# Realizamos la unión de los conjuntos
countries_union = countries.union(countries2)
print('Unión usando método union():', countries_union)

# También podemos usar el operador | para realizar la unión
countries_union2 = countries | countries2
print('Unión usando operador |:', countries_union2)
```

```python
# 2. Intersección de conjuntos

# La intersección obtiene los elementos que están presentes en ambos conjuntos
# Se puede realizar con el operador & o el método intersection()

# Definimos dos conjuntos
set_a = {'col', 'mex', 'bol', 'per'}  
set_b = {'per', 'bol', 'bra'}

# Realizamos la intersección usando el método intersection()
intersection_result = set_a.intersection(set_b)
print('Intersección usando método:', intersection_result)

# También podemos usar el operador & 
intersection_result2 = set_a & set_b
print('Intersección usando operador &:', intersection_result2)
```

```python
# 3. Diferencia de conjuntos

# La diferencia obtiene los elementos que están en el primer conjunto pero no en el segundo
# Se puede realizar con el operador - o el método difference()

# Definimos dos conjuntos
set_a = {'col', 'mex', 'bol', 'per'}
set_b = {'per', 'bol', 'bra'}

# Realizamos la diferencia usando el método difference()
difference_result = set_a.difference(set_b)
print('Diferencia usando método:', difference_result)

# También podemos usar el operador -
difference_result2 = set_a - set_b 
print('Diferencia usando operador -:', difference_result2)
```

```python
# 4. Diferencia simétrica de conjuntos

# La diferencia simétrica obtiene los elementos que están en uno u otro conjunto, pero no en ambos
# Se puede realizar con el operador ^ o el método symmetric_difference()

# Definimos dos conjuntos
set_a = {'col', 'mex', 'bol', 'per'}
set_b = {'per', 'bol', 'bra'}

# Realizamos la diferencia simétrica usando el método symmetric_difference()
symmetric_difference_result = set_a.symmetric_difference(set_b)
print('Diferencia simétrica usando método:', symmetric_difference_result)

# También podemos usar el operador ^
symmetric_difference_result2 = set_a ^ set_b
print('Diferencia simétrica usando operador ^:', symmetric_difference_result2)
```

```python
# Refactorizando el código.
import random

def choose_options():
  options = ('piedra', 'papel', 'tijera')
  user_option = random.choice(options)
  computer_option = random.choice(options)

  if not user_option in options:
    print('esa opcion no es valida')
    return None, None
  
  return user_option, computer_option


def check_rules(user_option, computer_option, user_wins, computer_wins):
    """
    Optimized version of check_rules function for rock-paper-scissors game.
    Uses a dictionary to define winning relationships instead of nested conditionals.
    """
    if user_option == computer_option:
        print('Empate!')
        return user_wins, computer_wins
    
    # Define winning relationships
    winning_rules = {
        'piedra': 'tijera',
        'papel': 'piedra',
        'tijera': 'papel'
    }
    
    if winning_rules[user_option] == computer_option:
        print(f'{user_option} gana a {computer_option}')
        print('user gano!')
        user_wins += 1
    else:
        print(f'{computer_option} gana a {user_option}')
        print('computer gano!')
        computer_wins += 1
        
    return user_wins, computer_wins 

def run_game():
    # Define constants
    MAX_WINS = 2
    
    # Initialize game state
    game_state = {
        'computer_wins': 0,
        'user_wins': 0,
        'rounds': 1
    }
    
    while True:
        # Display round information
        print('*' * 10)
        print(f"ROUND {game_state['rounds']}")
        print('*' * 10)
        print(f"computer_wins: {game_state['computer_wins']}")
        print(f"user_wins: {game_state['user_wins']}")
        
        # Increment round counter
        game_state['rounds'] += 1
        
        # Get options and determine winner
        user_option, computer_option = choose_options()
        game_state['user_wins'], game_state['computer_wins'] = check_rules(
            user_option, 
            computer_option, 
            game_state['user_wins'], 
            game_state['computer_wins']
        )
        
        # Check for winner
        if game_state['computer_wins'] >= MAX_WINS or game_state['user_wins'] >= MAX_WINS:
            winner = 'la computadora' if game_state['computer_wins'] >= MAX_WINS else 'el usuario'
            print(f'El ganador es {winner}')
            break 

run_game()
```

### HOF - Higher order functions: una funcion dentro de otra funcion.
<img src="./images/14-Hight-order-functions.jpg" alt="Scope" width="30%" />

```python
# Creando una funcion de tipo Higgh order function.
def increment(x):
    return x + 1

def hight_order_function(x, func):
    return x + func(x)

result = hight_order_function(2,increment)

# 2 + (2 + 1)
print(result)
```

```python
# Creando una funcion de tipo Higgh order function con lambda functions.

increment_v2 = lambda x: x + 1
hight_order_function_v2 = lambda x, func : x + func(x) 

result = hight_order_function_v2(2, increment_v2)
print(result)
```

```python
# Transformando a una lista en formato diccionario, la llave del diccionario sera el nombre de la columna.
# Importar y leer archivos CSV.
import csv

def read_csv(path):
    with open(path, 'r') as csvfile:
        reader = csv.reader(csvfile,  delimiter=',')
        header = next(reader)
        data = []
        
        for row in reader:
            iterable = zip(header,row)
            country_dict = {key: value for key, value in iterable}
            data.append(country_dict)
        return data

if __name__ == '__main__':
    data = read_csv('./inputs/world_population.csv')
    print(data)
```

```python
# Creando graficas con Matplotlib.
# %pip install matplotlib
import matplotlib.pyplot as plt

def generate_bar_chart(labels, values):
    fig, ax = plt.subplots()
    ax.bar(labels, values)
    plt.show()

def generate_pie_chart(labels, values):
    fig, ax = plt.subplots()
    ax.pie(values, labels=labels)
    ax.axis('equal')
    plt.show()

if __name__ == '__main__':
    labels = ['a','b','c']
    values = [100, 200, 80]
    generate_bar_chart(labels, values)
    generate_pie_chart(labels, values)

```

```python
import matplotlib.pyplot as plt

def get_population_by_country(data, country):
    """
    Find and return data for a specific country
    """
    result = list(filter(lambda item: item['Country/Territory'] == country, data))
    return result[0] if result else None

def population_chart(country_data):
    """
    Generate a chart of population growth for a specific country
    """
    if not country_data:
        return "Country not found"
    
    # Extract years and corresponding populations
    years = ['2022', '2020', '2015', '2010', '2000', '1990', '1980', '1970']
    population_values = [int(country_data[f'{year} Population']) for year in years]
    
    # Reverse lists to show chronological order
    years.reverse()
    population_values.reverse()
    
    # Create bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(years, population_values, color='skyblue')
    plt.title(f"Population Growth of {country_data['Country/Territory']}")
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.ticklabel_format(style='plain', axis='y')  # Avoid scientific notation
    
    # Add the population values on top of each bar
    for i, value in enumerate(population_values):
        plt.text(i, value + (max(population_values) * 0.02), 
                 f"{value:,}", 
                 ha='center', 
                 fontsize=9)
    
    plt.tight_layout()
    return plt

# Get country from user
country_name = input("Enter country name: ")

# Get data for that country
country_data = get_population_by_country(data, country_name)

if country_data:
    chart = population_chart(country_data)
    chart.show()
else:
    print(f"Country '{country_name}' not found in the dataset.")
```
